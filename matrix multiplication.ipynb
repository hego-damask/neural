```python
class Matrix:
    def __init__(self, matrix):
        self.matrix = matrix
        self.shape = (len(matrix), len(matrix[0]))
        self.T = [[0 for i in range(self.shape[0])] for j in range(self.shape[1])]
        for i in range(self.shape[1]):
            for j in range(self.shape[0]):
                self.T[i][j] = self.matrix[j][i]
    def dot(self, other):
        if type(other) != type(self):
            other = Matrix(other)
        if self.shape[-1] != other.shape[0]:
            print('error')
            raise Exception('Shape mismatch')
        else:
            m1 = (self.matrix[0][0]+self.matrix[1][1])*(other.matrix[0][0]+other.matrix[1][1])
            m2 = (self.matrix[1][0]+self.matrix[1][1])*other.matrix[0][0]
            m3 = self.matrix[0][0]*(other.matrix[0][1] - other.matrix[1][1])
            m4 = self.matrix[1][1]*(other.matrix[1][0] - other.matrix[0][0])
            m5 = (self.matrix[0][0]+self.matrix[0][1])*other.matrix[1][1]
            m6 = (self.matrix[1][0]-self.matrix[0][0])*(other.matrix[0][0]+other.matrix[0][1])
            m7 = (self.matrix[0][1]-self.matrix[1][1])*(other.matrix[1][0]+other.matrix[1][1])
            c = [[0, 0], [0, 0]]
            c[0][0] = m1 + m4 - m5 + m7
            c[0][1] = m3 + m5
            c[1][0] = m2 + m4
            c[1][1] = m1 - m2 + m3 + m6
            return Matrix(c)
```
