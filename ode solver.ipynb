```python
import numpy as np
from matplotlib import pyplot as plt
from scipy.integrate import odeint
```

```python
def Y(t):
    #return 5*np.exp(-2*t)
    return np.sin(t)
```

```python
def f(t, y = None):
    #return -10*np.exp(-2*t)
    return np.cos(t)
```

```python
h = 0.01
t = np.arange(0, 10, h)
y = []
yn = f(t[0])
y.append(yn)
for i in t[1:]:
    yn = yn + h*f(i)
    y.append(yn)

sol = odeint(f, 1.0, t)
```

```python
#plt.plot(t, f(t), label = 'dy/dt')
plt.plot(t, Y(t), label = 'y')
plt.plot(t, f(t), label = 'dy/dt')
plt.plot(t, y, label = 'euler')
plt.plot(t, sol, label='scipy')
plt.legend()
plt.grid(True)
```
