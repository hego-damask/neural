```python
from neural import Tensor
import numpy as np
```

```python
x = Tensor([1,2,3])
w = Tensor(size=3)
b = Tensor(size=1)
z = Tensor.sigmoid(w.dot(x)+b)
print(w)
print(b)
print(z)
```

```python
np_sigmoid = lambda x: 1/(1+np.exp(-x))
xnp = np.array(x.data)
wnp = np.array(w.data)
bnp = np.array(b.data)
znp = np_sigmoid(np.array(wnp.dot(xnp)+bnp))
print(xnp)
print(wnp)
print(bnp)
print(znp)
```
